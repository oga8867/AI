```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color = 'blue') #색상의 이름으로 지정
plt.plot(x, np.sin(x-1), color = 'g') #짧은 색상코드
plt.plot(x, np.sin(x-2), color = '0.75') #0과 1사이의 회색조를 표현한다.
plt.plot(x, np.sin(x-3), color = '#FFDD44') #16진수로 표현한 RGB
plt.plot(x, np.sin(x-4), color = 1.0, 0.2, 0.3) #rgb값을 튜플로 표현
plt.plot(x, np.sin(x-5), color = 'chartreuse')
```

```python
plt.plot(x, x+0, linestyle = 'solid')
plt.plot(x, x+1, linestyle = 'dotted')
plt.plot(x, x+2, linestyle = 'dashed')
plt.plot(x, x+3, linestyle = 'dashdot')
plt.plot(x, x+4, linestyle = '-')

plt.plot(x, x+5, linestyle = '--')
plt.plot(x, x+6, linestyle = ':')
```

```python
plt.plot(x, x+0,'-g')
plt.plot(x, x+1,'--c')
plt.plot(x, x+2,'-,k')
plt.plot(x, x+3,':r')
```

```python
#플롯 조정하기
plt.plot(x, np.sin(x))

plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]) #앞에 두개가 x축 뒤에가 y축 최소치 최대치
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
#플롯에 레이블 붙이기
plt.plot(x, np.sin(x))
plt.title('a sin curve')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin')
plt.plot(x, np.cos(x), ':b', label='cos')
plt.axis('equal')
plt.legend()
```

```python
# 산점도의 출력
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o',',','x','+','v','<','>','s','d']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
    plt.legend()
    plt.xlim(0,1,8)
```
