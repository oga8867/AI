```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random
plt.style.use('seaborn-whitegrid')
```

```python
def liner_function(x):
    a = 0.5 #기울기
    b = 2 #y절편
    
    return(a*x+b)

liner_function(5)
```

```python
x = np.arange(-10,10,0.1)
```

```python
y = liner_function(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('liner')
```

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    return a*x**2+b*x+c #ax^2+bx+c 이라는 2차함수

y = quadratic_function(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 3차 함수 y=ax^3 + bx^2 + cx + d

def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a*x**3+b*x**2+c*x+d

y  = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('cubic_function')
```

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    
    return a*x**2+b+x+c

y = my_func(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('my_func')
plt.scatter(1.5, my_func(1.5))
```
