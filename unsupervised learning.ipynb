```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
wine = load_wine()
```

```python
print(wine.DESCR)
```

```python
data = wine.data
label = wine.target
columns = wine.feature_names

data = pd.DataFrame(data, columns=columns)
data.head()
```

```python
# k-Means

# 데이터 전처리
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
data.shape
```

```python
# PCA (차원의 축소)
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
data = pca.fit_transform(data)
```

```python
data
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(data)
```

```python
cluster = kmeans.predict(data)
```

```python
type(cluster)
```

```python
plt.scatter(data[:,0],data[:,1], c=cluster,
            edgecolor='black',linewidth=1)
```

```python
#hierarchical Clustering 계층적 클러스터링
#디스턴스 구하는 3가지방법
from sklearn.cluster import AgglomerativeClustering
single_clustering = AgglomerativeClustering(n_clusters=3, linkage='single')
complete_clustering = AgglomerativeClustering(n_clusters=3, linkage='complete')
average_clustering = AgglomerativeClustering(n_clusters=3, linkage='average')

single_clustering.fit(data)
complete_clustering.fit(data)
average_clustering.fit(data)
```

```python
single_clustering
```

```python
single_cluster = single_clustering.labels_
complete_cluster = complete_clustering.labels_
average_cluster = average_clustering.labels_
```

```python
from scipy.cluster.hierarchy import dendrogram #덴드로그램
plt.figure(figsize=(20,20))

children = single_clustering.children_
```

```python
print(children)
```

```python
distance = np.arange(children.shape[0])
distance
```

```python
#전체 노드 수 계산
number_of_observations = np.arange(2,children.shape[0]+2)
number_of_observations
```

```python
linkage_matrix = np.column_stack([children, distance, number_of_observations]).astype(float)
```

```python
dendrogram(linkage_matrix, p=len(data), labels=single_cluster,
           show_contracted=True, no_labels=True)
```

```python
#silhoutette 실루엣으로 점수매기기, 비지도학습 평가

from sklearn.metrics import silhouette_score
best_n = 1
best_score = -1

for n_cluster in range(2,11):
    kmeans = KMeans(n_clusters=n_cluster) #KMeans 알고리즘에 파라미터를 n_cluster값으로 넣어줌
    kmeans.fit(data)
    cluster=kmeans.predict(data)
    score = silhouette_score(data, cluster)
    print('클러스터의 수:{} 실루엣 점수 : {:.4f}'.format(n_cluster,score))
    
    if score > best_score:
        best_n = n_cluster
        best_score = score
        
print('가장 높은 실루엣 점수를 가진 클러스터 수:{}, 실루엣 점수{:.4f}'.format(best_n,best_score))
```

```python
from sklearn.metrics import silhouette_score
best_n = 1
best_score = -1

for n_cluster in range(2,11):
    average_clustering = AgglomerativeClustering(n_clusters=n_cluster,linkage='average') #위에 계층적 클러스터 써봄
    average_clustering.fit(data)
    cluster=average_clustering.labels_ #predict 가 아니라 fit 해야함 predict 가 없음
    score = silhouette_score(data, cluster)
    print('클러스터의 수:{} 실루엣 점수 : {:.4f}'.format(n_cluster,score))
    
    if score > best_score:
        best_n = n_cluster
        best_score = score
        
print('가장 높은 실루엣 점수를 가진 클러스터 수:{}, 실루엣 점수{:.4f}'.format(best_n,best_score))
```
