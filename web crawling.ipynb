```python
import bs4
import requests
url = 'https://search.shopping.naver.com/search/all?query=%EB%85%B8%ED%8A%B8%EB%B6%81&cat_id=&frm=NVSHATC'
result = requests.get(url).text
result
```

```python
baObj = bs4.BeautifulSoup(result, 'html.parser')
```

```python
baObj.find('a',{'class':"basicList_link__JLQJf"}).text
```

```python
items = baObj.find_all('a',{'class':"basicList_link__JLQJf"})
```

```python
for item in items:
    print(item.text)
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```

```python
lists = items[:3]
for l in lists:
    print(l.text)
```
