```python
#데이터의 준비
from sklearn.model_selection import train_test_split
```

```python
train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

```python
x_train, x_test, y_train,y_test = train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0,train_size=0.8)

# x는 속성들 y는 꽃 종류  test_size는 데이터의 트레인과 테스트의 비율을 말함 train_size는 데이터의 트레인데이터를 바꿈 사실 앞에것과 같음
```

```python
print('x_train 크기:',x_train.shape)
print('y_train 크기:',y_train.shape)
print('x_test 크기:',x_test.shape)
print('y_test 크기:',y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
#예측하기

import numpy as np
x_new = np.array([[5,2.9,1,0.2]]) #예측용 샘플 데이터
x_new
```

```python
#모델의 평가
y_pred = knn.predict(x_test)
print("테스트 셋에 대한 예측값:\n",y_pred)
```

```python
np.mean(y_pred == y_test)  #정확도
```
